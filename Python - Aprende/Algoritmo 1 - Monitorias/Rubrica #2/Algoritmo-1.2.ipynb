```python

print("Lista de Productos: \nA. Producto-A \nB. Producto-B \nC. Producto-C")

while True:
    opcion2 = 0  # Variable para controlar la repetición del segundo menú
    print("**Se sugiere digitar su seleccion en Mayuscula**")
    opcion = input("Seleccione un producto (A, B, C): ")

    if opcion in ['A', 'B', 'C']:
        if opcion == 'A':
            while opcion2 == 0:
                opcion2 = int(input("Seleccione un adicional: \n1. Envase de vidrio \n2.Etiquetas de advertencia \n3.Sensor de Temperatura \nSeleccione: "))

                if opcion2 in [1, 2, 3]:
                    Adicional = ["Envase de vidrio", "Etiquetas de advertencia", "Sensor de Temperatura"]
                    print(f"\nEl producto seleccionado es Producto-A y el adicional elegido es: {Adicional[opcion2-1]}")
                    break
                else:
                    print("Opción no válida. Por favor, seleccione un número entre 1 y 3.")
                    opcion2 = 0  # Reinicia el controlador para volver a pedir una opción valida

        elif opcion == 'B':
            while opcion2 == 0:
                opcion2 = int(input("Seleccione un adicional: \n1. Barriles de metal \n2. Empaque sellado \n3. Alarma de seguridad \nSeleccione: "))

                if opcion2 in [1, 2, 3]:
                    Adicional = ["Barriles de metal", "Empaque sellado", "Alarma de seguridad"]
                    print(f"\nEl producto seleccionado es Producto-B y el adicional elegido es: {Adicional[opcion2-1]}")
                    break
                else:
                    print("Opción no válida. Por favor, seleccione un número entre 1 y 3.")
                    opcion2 = 0  # Reinicia el controlador para volver a pedir una opción valida
        else:
            while opcion2 == 0:
                opcion2 = int(input("Seleccione un adicional: \n1. Contenedores plasticos \n2. Control de Humedad \n3. Cámara de Frio \nSeleccione: "))

                if opcion2 in [1, 2, 3]:
                    Adicional = ["Contenedores plasticos", "Control de Humedad", "Cámara de Frio"]
                    print(f"\nEl producto seleccionado es Producto-C y el adicional elegido es: {Adicional[opcion2-1]}")
                    break
                else:
                    print("Opción no válida. Por favor, seleccione un número entre 1 y 3.")
                    opcion2 = 0  # Reinicia el controlador para volver a pedir una opción valida        
    
    else:
        print("Opción no válida. Por favor, seleccione A, B o C.")

```

```python
Swich = True

while Swich == True:
    Horas = int(input("Ingrese numero de horas de practica: "))

    if Horas >= 0 and Horas <= 120:
        if Horas >= 0 and Horas <= 30: #Hasta 30 Horas
            fecha = "15 de enero"
        elif Horas > 30 and Horas < 60: # Mas de 30 y menos de 60
            fecha = "15 de febrero"
        elif Horas >= 60 and Horas <= 90: # Entre 60 y 90
            fecha = "15 de marzo"
        else: # mas de 90
            fecha = "15 de abril"
        print(f"La fecha es: {fecha}")
    else:
        print("Error: Horas no válidas. Ingrese un número entre 0 y 120.")

    while True:
        opcion = input("¿Desea continuar? (si/no): ")
        if opcion == "si" or opcion == "no":
            break
        else:
            print("Error: Opción no válida. Ingrese 'si' o 'no'.")

    if opcion == "si":
        Swich = True
    elif opcion == "no":
        print("Gracias por utilizar el programa.")
        Swich = False    



```
